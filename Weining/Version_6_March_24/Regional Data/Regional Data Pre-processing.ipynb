```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

# Part I: Nodes 

```python
Nodes = pd.read_csv('CountryNodes_All.csv')
Nodes.head(1)
```

```python
Nodes[Nodes['Region'] == 'Northern America']
```

```python
# Calculate influenza case number by 22 region

# create a df
RegionNameList = Nodes['Region'].unique()
data = {'Region': RegionNameList}
df = pd.DataFrame(data)

# use group by to caculate influenza data
a = pd.DataFrame(Nodes.groupby('Region')['Influenza2005'].sum()).reset_index()

# merge a with df based on df's Region column 
df = pd.merge(df, a, on = 'Region')

# do the same for year 2010
Y2010 = pd.DataFrame(Nodes.groupby('Region')['Influenza2010'].sum()).reset_index()
df = pd.merge(df, Y2010, on = 'Region')

Y2015 = pd.DataFrame(Nodes.groupby('Region')['Influenza2015'].sum()).reset_index()
df = pd.merge(df, Y2015, on = 'Region')

Y2019 = pd.DataFrame(Nodes.groupby('Region')['Influenza2019'].sum()).reset_index()
df = pd.merge(df, Y2019, on = 'Region')

df
```

```python
# create a second region feature as the node color
list_Colr = ['Oceania', 'Latin America and The Caribbean', 'Latin America and The Caribbean',
            'Central and Southern Asia', 'Sub_Saharan Africa', 'Eastern and South Eastern Asia',
            'Europe and Northern America', 'Oceania', 'Oceania',
            'Sub_Saharan Africa', 'Northern Africa and Western Asia', 'Europe and Northern America',
            'Europe and Northern America', 'Oceania', 'Latin America and The Caribbean',
            'Eastern and South Eastern Asia', 'Sub_Saharan Africa', 'Central and Southern Asia',
            'Europe and Northern America', 'Sub_Saharan Africa', 'Northern Africa and Western Asia',
            'Europe and Northern America']

# merge with df
data = {'UN_Region': list_Colr}
df_colr = pd.DataFrame(data)

df['UN_Region'] = df_colr['UN_Region']
df.head()
```

```python
df.to_csv('Nodes_Regional_Data.csv', index=False)
```

# Part II: Edgelist

```python
Links = pd.read_csv('EdgeList_All.csv')
Links = Links[['Destination', 'Origin', 'flow2000', 'flow2005', 'flow2010', 'flow2015', 'flow2019']]
Links.head()
```

```python
df_2 = Nodes[['Country', 'Region']].rename(columns = {'Country': 'Destination', 'Region': 'Destination_Region'})
df_2.head()
```

```python
# merge links and df_2 together based on Country name
Links = pd.merge(Links, df_2, on = 'Destination', how = 'left')
Links.head(20)
```

```python
df_3 = Nodes[['Country', 'Region']].rename(columns = {'Country': 'Origin', 'Region': 'Origin_Region'})
df_3.head()
```

```python
# merge links and df_3 together based on Country name
Links = pd.merge(Links, df_3, on = 'Origin', how = 'left')
Links.head(20)
```

```python
# create a df to store the data
# create a df
RegionNameList = Nodes['Region'].unique()
list1 = [ele for ele in RegionNameList for i in range(len(RegionNameList))] 
list2 = list(RegionNameList) * len(RegionNameList)

data = {'Destination_Region': list1, "Origin_Region": list2}
df = pd.DataFrame(data)
df.head()
```

```python
len(df)
```

```python
# Calculate migration flow based on region name
# use group by to caculate influenza data
Y2000 = pd.DataFrame(Links.groupby(['Destination_Region', 'Origin_Region'])['flow2000'].sum()).reset_index()

# merge a with df based on df's Region column 
df = pd.merge(df, Y2000, how = 'left', left_on = ['Destination_Region', 'Origin_Region'], 
              right_on = ['Destination_Region', 'Origin_Region'])

# do the same for year 2005
Y2005 = pd.DataFrame(Links.groupby(['Destination_Region', 'Origin_Region'])['flow2005'].sum()).reset_index()

# merge a with df based on df's Region column 
df = pd.merge(df, Y2005, how = 'left', left_on = ['Destination_Region', 'Origin_Region'], 
              right_on = ['Destination_Region', 'Origin_Region'])

df.head()
```

```python
# do the same for year 2010
Y2010 = pd.DataFrame(Links.groupby(['Destination_Region', 'Origin_Region'])['flow2010'].sum()).reset_index()

# merge a with df based on df's Region column 
df = pd.merge(df, Y2010, how = 'left', left_on = ['Destination_Region', 'Origin_Region'], 
              right_on = ['Destination_Region', 'Origin_Region'])

# do the same for year 2015
Y2015 = pd.DataFrame(Links.groupby(['Destination_Region', 'Origin_Region'])['flow2015'].sum()).reset_index()

# merge a with df based on df's Region column 
df = pd.merge(df, Y2015, how = 'left', left_on = ['Destination_Region', 'Origin_Region'], 
              right_on = ['Destination_Region', 'Origin_Region'])

# do the same for year 2019
Y2019 = pd.DataFrame(Links.groupby(['Destination_Region', 'Origin_Region'])['flow2019'].sum()).reset_index()

# merge a with df based on df's Region column 
df = pd.merge(df, Y2019, how = 'left', left_on = ['Destination_Region', 'Origin_Region'], 
              right_on = ['Destination_Region', 'Origin_Region'])

df.head()
```

## Calculate increase rate based on regional data 

```python
# Calcualte migration flow increase rate
df['increase_rate_2000_2005'] = 100 * ((df['flow2005'] - df['flow2000'])/df['flow2000'])
df['increase_rate_2000_2005'] = df['increase_rate_2000_2005'].fillna(0)
df.loc[~np.isfinite(df['increase_rate_2000_2005']), 'increase_rate_2000_2005'] = float(0)
df
```

```python
# Calcualte migration flow increase rate
df['increase_rate_2005_2010'] = 100 * ((df['flow2010'] - df['flow2005'])/df['flow2005'])
df['increase_rate_2005_2010'] = df['increase_rate_2005_2010'].fillna(0)
df.loc[~np.isfinite(df['increase_rate_2005_2010']), 'increase_rate_2005_2010'] = float(0)
df.head()
```

```python
# Calcualte migration flow increase rate
df['increase_rate_2010_2015'] = 100 * ((df['flow2015'] - df['flow2010'])/df['flow2010'])
df['increase_rate_2010_2015'] = df['increase_rate_2010_2015'].fillna(0)
df.loc[~np.isfinite(df['increase_rate_2010_2015']), 'increase_rate_2010_2015'] = float(0)
df.head()
```

```python
# Calcualte migration flow increase rate
df['increase_rate_2015_2019'] = 100 * ((df['flow2019'] - df['flow2015'])/df['flow2015'])
df['increase_rate_2015_2019'] = df['increase_rate_2015_2019'].fillna(0)
df.loc[~np.isfinite(df['increase_rate_2015_2019']), 'increase_rate_2015_2019'] = float(0)
df.head()
```
